```python
import kagglehub
path = kagglehub.dataset_download("uciml/student-alcohol-consumption")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv(path+"/student-mat.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.corr(numeric_only=True)["G3"].sort_values(ascending=False)
```

```python
df=df.drop(['school',"address","famsize","Pstatus","Mjob","Fjob","guardian","reason","Dalc","Walc"],axis=1)
```

```python
df.shape
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
for col in df.select_dtypes(exclude=np.number).columns:
  df[col]=le.fit_transform(df[col])
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.iloc[:, :-1] #all rows,all cols (-1->except last one)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=scaler.fit_transform(X)
```

```python
X
```

```python
Y=df.iloc[:,-1] #all row and only last column/
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LR_model=LinearRegression()
LR_model.fit(X_train,Y_train)
```

```python
LR_model.predict(X_test)  #predicted result on hidden(Test data)
```

```python
Y_test  #actual result
```

```python
from sklearn.metrics import accuracy_score ,confusion_matrix,r2_score,mean_squared_error
```

```python
LR_R2_Score=r2_score(Y_test,LR_model.predict(X_test))
LR_MSE=mean_squared_error(Y_test,LR_model.predict(X_test))
LR_RMSE=np.sqrt(LR_MSE)

print("LR_R2_Score:",LR_R2_Score)
print("LR_MSE:",LR_MSE)
print("LR_RMSE:",LR_RMSE)
```

```python
plt.scatter(Y_test,LR_model.predict(X_test))
plt.plot(Y_test,Y_test,color="red")
plt.xlabel('Actual Result'),
plt.ylabel("Predicted Result")
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
RF_model=RandomForestRegressor()
RF_model.fit(X_train,Y_train)
```

```python
RF_model.predict(X_test)
```

```python
Y_test
```

```python
RF_R2_Score=r2_score(Y_test,RF_model.predict(X_test))
RF_MSE=mean_squared_error(Y_test,RF_model.predict(X_test))
RF_RMSE=np.sqrt(RF_MSE)


print("RF_R2_Score:",RF_R2_Score)
print("RF_MSE:",RF_MSE)
print("RF_RMSE:",RF_RMSE)
```

```python
plt.scatter(Y_test,RF_model.predict(X_test))
plt.plot(Y_test,Y_test,color="red")
plt.xlabel('Actual Result'),
plt.ylabel("Predicted Result")
plt.show()
```

```python
comparison_table = pd.DataFrame({
    "Metric": ["R2 Score", "MSE", "RMSE"],
    "Linear Regression": [LR_R2_Score, LR_MSE, LR_RMSE],
    "Random Forest": [RF_R2_Score, RF_MSE, RF_RMSE]
})

print("\n📊 Model Comparison Table:\n")
print(comparison_table)
```
